```python
data = pd.DataFrame({'Months': ['jan','feb','mar','apr','may','jun','jul','agu','sep','oct','nov','dec'],
                    'Sales': [400,500,700,450,620,400,500,700,450,620,456,625],
                     'Profit': [220,78,392,35,62,220,78,392,235,162,135,285],
                    })
```

```python
data
```

```python
fig, ax = plt.subplots(figsize = (10,3))
ax.bar(data['Months'],data['Sales'],color = 'red',label= 'Sales')
ax.plot(data['Months'],data['Profit'],color = 'white',label= 'Profit')

ax.set(xlabel = 'Months', ylabel = 'Amount in inr', title = 'Sales VS Profit charts')
ax.legend(loc = 'upper left')
plt.grid(True)
plt.show()
```

```python
Sales = [100,200,300,400,500]
Quntity = [10,20,30,40,50]
Profit = [90,60,40,50,20]

fig, ax = plt.subplots(figsize = (8,4))
ax.bar(np.arange(len(Sales)),Sales, align = 'center', color = 'black')

# plot line chart for quntity
ax.plot(np.arange(len(Quntity)), Quntity, color = 'red', linestyle= 'dashed')

# plot line chart for profit
ax.plot(np.arange(len(Profit)), Profit, color = 'white', linestyle = 'dotted')

# adding labels and annotation
ax.set(xlabel = 'Months', ylabel = 'Amount in inr', title = 'Sales VS Profit charts')
ax.set_xticks(np.arange(len(Sales)))
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May'])
ax.legend(['Quantity','Profit'])

# Adding a data labels
for i, v in enumerate(Sales):
    ax.text(i,v,str(v),ha= 'center', va = 'bottom')

plt.show()
```

# create coloumn chart and horizontal bar chart together

```python
x = np.array(['germany','usa','uk','france','mexico'])
y1 = np.array([10,20,30,40,50])
y2 = np.array([24,52,63,20,74])


fig, ax = plt.subplots(2,1)
ax[0].bar(x,y1, color=['black','red','black','red','black'])
ax[0].set_title('bar chart')

ax[1].barh(x,y2, color=['black','red','black','red','black'])
ax[1].set_title('horizontal bar chart')
plt.grid(True)
ax[0].grid(True)
plt.style.use('seaborn-v0_8-dark')
```

```python
x = np.array(['germany','usa','uk','france','mexico'])
y1 = np.array([10,20,30,40,50])
y2 = np.array([24,52,63,20,74])


fig, ax = plt.subplots(2,1,figsize=(4,8))
ax[0].bar(x,y1, color=['black','red','black','red','black'])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('rate')
ax[0].set_title('bar chart')

ax[1].barh(x,y2, color=['black','red','black','red','black'])
ax[1].set_title('horizontal bar chart')
ax[1].set_xlabel('Country')
ax[1].set_ylabel('rate')
plt.grid(True)
ax[0].grid(True)
plt.style.use('seaborn-v0_8-dark')
#plt.tight_layout()
```

```python
x = np.array(['germany','usa','uk','france','mexico'])
y1 = np.array([10,20,30,40,50])
y2 = np.array([24,52,63,20,74])


fig, ax = plt.subplots(1,2,figsize = (12,4))
ax[0].bar(x,y1, color=['black','red','black','red','black'])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('rate')
ax[0].set_title('bar chart')

ax[1].barh(x,y2, color=['black','red','black','red','black'])
ax[1].set_title('horizontal bar chart')
ax[1].set_xlabel('rate')
ax[1].set_ylabel('Country')
plt.grid(True)
ax[0].grid(True)
plt.style.use('seaborn-v0_8-dark')
```

```python
x = np.array(['germany','usa','uk','france','mexico'])
y1 = np.random.randint(1000,9000,5)
y2 = np.random.randint(1000,9000,5)

bar_width = 0.40
r1 = np.arange(len(x))
r2 = [x + bar_width for x in r1]

fig, ax = plt.subplots(figsize = (6,4))
ax.bar(r1,y1,color = 'yellow',width = bar_width, edgecolor = 'black', label = 'y1')
ax.bar(r2,y2,color = 'red',width = bar_width, edgecolor = 'black', label = 'y2')

ax.set_xlabel('Country')
ax.set_ylabel('Sale')
ax.set_title('Chart of country')
ax.legend()
ax.grid(True)

plt.show()
```

```python
x = ['germany','usa','uk','france','mexico']
values = [50,54,78,48,51]
errors = [7,9,0.5,5,1]

fig, ax = plt.subplots(figsize = (4,4))
ax.bar(x,values,yerr = errors, align ='center', alpha = 0.5, ecolor = 'black', capsize = 10 )

ax.set(xlabel = 'Country', ylabel = 'Sale', title = 'Error bar')
```
